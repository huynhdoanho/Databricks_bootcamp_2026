```python
import pyspark.sql.functions as F
```

```python
df = spark.read.table("workspace.bronze.erp_location")
display(df.printSchema())
df.display()
```

```python
# trimming
df = df.withColumn('CNTRY', F.trim(F.col('CNTRY')))\
    .withColumn('CID', F.trim(F.col('CID')))
df.display()
```

```python
# rename columns 
df = df.withColumnRenamed('CNTRY', 'country')\
    .withColumnRenamed('CID', 'customer_id')
df.display()
```

```python
# normalize IDs
df = df.withColumn('customer_id', F.regexp_replace('customer_id', '-', ''))
df.display()
```

```python
# write table
df.write.mode("overwrite").saveAsTable("workspace.silver.erp_location")
```
